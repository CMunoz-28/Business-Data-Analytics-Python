```python
#Install library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
import statsmodels.api as sm
from causalinference import CausalModel
```

```python
#Load Data
df = pd.read_csv("stackoverflow.csv")
df.head()
```

#Data Analysis

```python
#Picking variables
df = df.iloc[:, :8]
df.head(0)
```

```python
#Transforming character variables
df = pd.get_dummies(df, drop_first= True)
df.head(1)
```

```python
#Looking at Group means
df.groupby("Remote_Remote").mean()
```

```python
#T-test
#stating which variables to test
continuous = ['Salary', 'YearsCodedJob']

#Where to store results
stat = {}
pvalue = {}

#Loop
for x in continuous:
  group1 = df.where(df.Remote_Remote == 0).dropna()[x]
  group2 = df.where(df.Remote_Remote == 1).dropna()[x]
  stat[x], pvalue[x] = ss.ttest_ind(group1, group2)
ttests = pd.DataFrame.from_dict(pvalue, orient = 'Index')
ttests.columns = ['pvalue']
print(ttests)
```

#Matching

```python
#Isolate y, treat, confounders
y = df.CareerSatisfaction.values
treat = df.Remote_Remote.values

# Convert boolean columns to integers before creating the confounders array
df_numeric = df.copy()
for col in ['Country_Germany', 'Country_India', 'Country_United Kingdom', 'Country_United States']:
  df_numeric[col] = df_numeric[col].astype(int)

confounders = df_numeric.drop(columns = ["Remote_Remote",
                                 "CareerSatisfaction"]).values
```

```python
#Matching
model = CausalModel(y, treat, confounders)
model.est_via_matching(bias_adj = True)
print(model.estimates)
```

#Robustness check

```python
#Remove 1 confounder
confounders = df_numeric.drop(columns = ["Remote_Remote",
                                 "CareerSatisfaction", "Hobby"]).values
```

```python
#Matching
model = CausalModel(y, treat, confounders)
model.est_via_matching(bias_adj = True)
print(model.estimates)
```
